```python
df["TotalCharges"] = df["TotalCharges"].isna().mean()
df
```

```python
df2 = df.copy()
category = np.empty(0)
for col in df2.columns:
    if df2[col].dtype=="object":
        category = np.append(category, col)
category
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}

for col in category:
    le = LabelEncoder()
    df2[col + "_encoded"] = le.fit_transform(df2[col])
    encoders[col] = le
df2.head()
```

```python


from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
for col in category:
    df2[col] = ohe.fit_transform(df2[[col]])
df2.head()
```

```python
num_col = np.empty(0)
for col in df2.columns:
    if df2[col].dtype=="int": 
        num_col = np.append(num_col, col)

from sklearn.preprocessing import StandardScaler

Std = StandardScaler()
for col in num_col:
    df2[col] = Std.fit_transform(df2[[col]])
df2.head()

```
